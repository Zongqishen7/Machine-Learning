```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, SGDRegressor,Ridge
from sklearn.metrics import mean_squared_error
```

```python
bt = load_boston()
bost = pd.DataFrame(data = bt.data, columns = bt.feature_names)
bost
```

```python
#线性回归
def linear1():
    """
    正规方程的优化方法对波士顿房价进行预测
    """
    #1)获取数据集
    bt = load_boston()
    #2)划分数据集
    X_train, X_test, y_train, y_test = train_test_split(bt.data, bt.target, random_state=22)
    #3)特征工程:
    transfer = StandardScaler()
    transfer.fit_transform(X_train)
    transfer.transform(X_test)
    #4)预估器流程 
    estimator = LinearRegression()
    estimator.fit(X_train, y_train)
    #5)模型评估
    preditced = estimator.predict(X_test)
    error = mean_squared_error(y_test, preditced)
    print("正规方程 --- 权重系数为\n", estimator.coef_)
    print("正规方程 --- 偏置为\n", estimator.intercept_)
    print("均方误差\n", error)

    return None

```

```python
linear1()
```

```python
#线性回归
def linear2():
    """
    对波士顿房价进行预测
    """
    #1)获取数据集
    bt = load_boston()
    #2)划分数据集
    X_train, X_test, y_train, y_test = train_test_split(bt.data, bt.target, random_state=22)
    #3)特征工程:
    transfer = StandardScaler()
    transfer.fit_transform(X_train)
    transfer.transform(X_test)
    #4)预估器流程
    estimator = SGDRegressor(learning_rate="constant",eta0=0.01, max_iter=10000,penalty="l1")
    estimator.fit(X_train, y_train)
    #5)模型评估
    preditced = estimator.predict(X_test)
    error = mean_squared_error(y_test, preditced)
    print("梯度下降 --- 权重系数为\n", estimator.coef_)
    print("梯度下降 --- 偏置为\n", estimator.intercept_)
    print("均方误差\n", error)
    return None
```

```python
linear2()
```

```python

def linear3():
    """
    岭回归对波士顿房价进行预测
    """
    #1)获取数据集
    bt = load_boston()
    #2)划分数据集
    X_train, X_test, y_train, y_test = train_test_split(bt.data, bt.target, random_state=22)
    #3)特征工程:
    transfer = StandardScaler()
    transfer.fit_transform(X_train)
    transfer.transform(X_test)
    #4)预估器流程
    estimator = Ridge(alpha=0.5, max_iter=10000)
    estimator.fit(X_train, y_train)
    #5)模型评估
    preditced = estimator.predict(X_test)
    preditced = estimator.predict(X_test)
    error = mean_squared_error(y_test, preditced)
    print("岭回归 --- 权重系数为\n", estimator.coef_)
    print("岭回归 --- 偏置为\n", estimator.intercept_)
    print("均方误差\n", error)
    return None
```

```python
linear3()
```

```python
# 保存模型岭回归
import joblib

"""
岭回归对波士顿房价进行预测
"""
#1)获取数据集
bt = load_boston()
#2)划分数据集
X_train, X_test, y_train, y_test = train_test_split(bt.data, bt.target, random_state=22)
#3)特征工程:
transfer = StandardScaler()
transfer.fit_transform(X_train)
transfer.transform(X_test)
#4)预估器流程
estimator = Ridge(alpha=0.5, max_iter=10000)
estimator.fit(X_train, y_train)
#5)保存模型
joblib.dump(estimator, "my_ridge.pkl")


```
