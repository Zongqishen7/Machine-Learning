```python
import pandas as pd 
from sklearn.linear_model import LogisticRegression
data = pd.read_csv("/Users/shenzongqi/Documents/Stevens_First_semester/FE541/Group_Project/breast.csv")
data.head()
```

1. 查看缺失值

```python
data.isnull().sum()
```

```python
# 用均值填充缺失值的位置
from statistics import mean
data["bareNuclei"].fillna(value=data["bareNuclei"].mean(), inplace=True)
```

```python
data.isnull().sum()
```

2. 划分数据集

```python
from sklearn.model_selection import train_test_split
```

```python
data
```

```python
#data.drop(columns="Unnamed: 0",inplace=True)
X = data.drop(columns="class")
y = data["class"]
X_train, X_test, y_train, y_test = train_test_split(X,y)
X_test
```

3. 特征工程

```python
from sklearn.preprocessing import StandardScaler
```

```python
transformer = StandardScaler()
X_train = transformer.fit_transform(X_train)
X_test = transformer.transform(X_test)
```

4. 导入预估器

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
estimator = LogisticRegression()
estimator.fit(X_train,y_train)
predicted = estimator.predict(X_test)
```

```python
print("Accuracy\n", np.mean(y_test == predicted))
print("Score\n", estimator.score(X_test, y_test))
```

查看精确率召回率与F1score

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, predicted, labels=[2,4], target_names=["良性","恶性"])
print(report)
```
