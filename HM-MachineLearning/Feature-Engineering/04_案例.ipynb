```python
import pandas as pd
import numpy as np
order_products = pd.read_csv("/Users/shenzongqi/Downloads/order_products__prior.csv")
products = pd.read_csv("/Users/shenzongqi/Downloads/products.csv")
orders = pd.read_csv("/Users/shenzongqi/Downloads/orders.csv")
aisles = pd.read_csv("/Users/shenzongqi/Downloads/aisles.csv")
order_products.head()
```

```python
aisles.head()
```

```python
orders.head()
```

```python
products.head()
```

```python
table1 = pd.merge(aisles, products, on="aisle_id")
table1
```

```python
table_2 = pd.merge(table1, order_products, on = 'product_id')
table_2
```

```python
table3 = pd.merge(table_2, orders, on = "order_id")
```

```python
table3
```

```python
table = pd.crosstab(table3["user_id"],table3["aisle"])
```

```python
table.head()
```

```python
data = table[:10000]
data 
```

# PCA 降维

```python
from sklearn.decomposition import PCA
```

```python
# 1) 实例化一个转换器类
transfer = PCA(n_components=0.95)
# 2  调用fit_transform
new_data = transfer.fit_transform(data)
```

```python
pd.DataFrame(new_data)
```
