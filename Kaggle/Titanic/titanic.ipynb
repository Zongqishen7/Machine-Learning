# <font color = Darkblue size = 6>Pre- processing <font>

```python
train.isnull().sum()
```

```python
train["Age"] = train["Age"].fillna(train["Age"].mean())
```

```python
train
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
train["Sex"] = le.fit_transform(train["Sex"])
```

```python
train
```

```python
#df = pd.get_dummies(train, columns=["Sex"])
```

```python
X_train = train.drop(["Survived"],axis= 1)
y_train = train["Survived"]
```

```python
X_train
```

```python
test = pd.read_csv("test.csv")
test.drop(["Name","Ticket","Cabin","Embarked"],axis = 1,inplace = True)
test["Sex"] = le.fit_transform(test["Sex"])
test
```

```python
test.isnull().sum()
```

```python
test["Age"] = test["Age"].fillna(test["Age"].mean())
test["Fare"] = test["Fare"].fillna(test["Fare"].mean())
```

```python
test
```

```python
X_test = test
```

# <font color = darkblue size = 6> Modelling <font>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
print('Coefficient:\n', lr.coef_)
print('\nIntercept:', lr.intercept_)
```

```python
predictions = lr.predict(X_test)
predictions
```

```python
y_test = pd.read_csv("gender_submission.csv")
y_test
```

```python
print('Accuracy Score:\n', accuracy_score(y_test["Survived"], predictions)*100, '%')
confusion_matrix(y_test["Survived"],predictions)
```

# <font color = Darkblue size = 6>Submit_Predictions<font>

```python
Submit_Predictions = y_test.drop(["Survived"], axis = 1)
Submit_Predictions["Survived"] = predictions
Submit_Predictions
```
